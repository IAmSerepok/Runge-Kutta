```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(t, v):
    x, y = v
    return np.array([
        np.sin(t), 
        np.cos(t)
    ])
```

```python
def runge_kutta(f, t0, v0, h):
    k1 = f(t0, v0)
    k2 = f(t0 + h / 2, v0 + k1 * h / 2)
    k3 = f(t0 + h / 2, v0 + k2 * h / 2)
    k4 = f(t0 + h, v0 + k3 * h)
    return v0 + h * (k1 + 2 * k2 + 2 * k3 + k4) / 6
```

```python
h = 0.01
t0 = 0
v0 = 1, 0
```

```python
T, V0, V1 = [], [], []

for _ in range(10000):
    T.append(t0)
    V0.append(v0[0])
    V1.append(v0[1])
    t0 += h
    v0 = runge_kutta(f, t0, v0, h)
```

```python
plt.figure(figsize=(10, 8))

plt.plot(V0, V1)

plt.show()
```
